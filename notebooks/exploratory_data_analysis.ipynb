 ### a) Cargar el dataset y examinar la estructura y el tipo de datos

```python
df = pd.read_csv(r"C:\Users\Paul Tandazo\Desktop\PSet2-dataM\data\raw\Airbnb_Data.csv") 
df.head()
```

## Se utilizarán las siguientes columnas:

```python
# Filtrar el data frame para mantener solo las columnas seleccionadas
columnas_utilizadas = [
    'log_price', 'property_type', 'room_type', 'accommodates', 'bathrooms', 'bed_type', 'cancellation_policy', 'cleaning_fee', 'latitude', 'longitude', 'review_scores_rating', 
    'bedrooms', 'beds', 'host_identity_verified', 'instant_bookable'
]
df = df[columnas_utilizadas]

# Verificar el resultado
df.head()
```

```python
# Información de los datos restantes
df.info()
```

```python
# Mapeo de datos t y f a booleanos

df['host_identity_verified'] = df['host_identity_verified'].map({'t': True, 'f': False})
df['instant_bookable'] = df['instant_bookable'].map({'t': True, 'f': False})
```

```python
df.head()
```

```python
# Saber cuántos nulos hay
df.isnull().sum()
```

```python
# Todos los host_identity_verified nulos son falsos
df['host_identity_verified'].fillna(False, inplace=True)

# Verificacion
df.isnull().sum()
```

```python
# Rellenar valores de review scores con la media
df['review_scores_rating'].fillna(df['review_scores_rating'].mean(), inplace=True)
# Verificacion
df.isnull().sum()
```

```python
# Tras analizar el dataset, se estima que deben tener al menos 1 baño.
df['bathrooms'].fillna(1, inplace=True)
# Verificacion
df.isnull().sum()
```

```python
# Se elimina todas las filas que no tengan cuarto ni cama
df.dropna(subset=['bedrooms', 'beds'], inplace=True)
df['beds'].fillna(1, inplace=True)
df['bedrooms'].fillna(1, inplace=True)
# Verficiacion 
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Se transforman los datos para que puedan ser analizados por el modelo ML.

label_encoder = LabelEncoder()
df['property_type'] = label_encoder.fit_transform(df['property_type'])
df['room_type'] = label_encoder.fit_transform(df['room_type'])
df['bed_type'] = label_encoder.fit_transform(df['bed_type'])
df['cancellation_policy'] = label_encoder.fit_transform(df['cancellation_policy'])
```

```python
df.info()
```

```python
df.head()
```

```python
df['bedrooms'] = df['bedrooms'].astype('int32')
df['bathrooms'] = df['bathrooms'].astype('int32')
df['beds'] = df['beds'].astype('int32')

df.info()
```

```python
df['cleaning_fee'] = df['cleaning_fee'].astype('int32')
df['host_identity_verified'] = df['host_identity_verified'].astype('int32')
df['instant_bookable'] = df['instant_bookable'].astype('int32')
df.head()
```

```python
output_path = r"C:\Users\Paul Tandazo\Desktop\PSet2-dataM\data\clean\cleaned_data.csv"
df.to_csv(output_path, index=False, encoding='utf-8')

print(f"CSV file saved successfully at: {output_path}")
```
