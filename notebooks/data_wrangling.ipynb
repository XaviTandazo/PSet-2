```python
import sys
sys.path.append('../src')
```

```python
from data_wrangling import load_data, filter_columns, map_boolean_values, fill_missing_values, encode_labels, convert_column_types, save_cleaned_data
```

```python
# Cargar los datos
df = load_data('../data/raw/Airbnb_Data.csv')
df.head()
df.info()
```

```python
# Filtrar columnas
df = filter_columns(df)
df.head()
```

```python
# Mapear valores booleanos
df = map_boolean_values(df)
df.info()
```

```python
# Rellenar valores faltantes
df = fill_missing_values(df)
df.info()
```

```python
# Codificar columnas categóricas
df = encode_labels(df)
df.head()
```

```python
# Convertir tipos de columnas
df = convert_column_types(df)
df.head()
```

```python

# Guardar los datos limpios
save_cleaned_data(df, '../data/clean/cleaned_data.csv')
```
