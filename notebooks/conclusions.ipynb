```python
import matplotlib.pyplot as plt
```

```python
modelos = ['Linear Regression (Normal Equation)', 'Linear Regression (SVD)', 'Polynomial Regression', 
           'Batch Gradient Descent', 'SGDRegressor', 'Lasso Regression', 'Ridge Regression']
mse_values = [0.2591627613278307, 0.25916276132783084, 0.0056728202281362446, 0.3566843080910934,
              0.2637288588446325, 0.5149366954573091, 0.2591608210105241]
r2_values = [0.49669201916022154, 0.49669201916022143, 0.9889830788958194, 0.30729994547529427,
             0.4878244129126097, -3.467752484720421e-05, 0.49669578736048614]


# Crear grfico para MSE
plt.figure(figsize=(12, 6))
plt.barh(modelos, mse_values, color='lightcoral')
plt.xlabel('MSE (Mean Squared Error)')
plt.title('Comparacin de MSE entre los Modelos')
plt.show()
# Crear grfico para R
plt.figure(figsize=(12, 6))
plt.barh(modelos, r2_values, color='lightblue')
plt.xlabel('R (Coeficiente de Determinacin)')
plt.title('Comparacin de R entre los Modelos')
plt.show()
```

# Anlisis de Modelos para la Prediccin

## 1. Regresin Lineal (Ecuacin Normal y SVD)
- **MSE**: 0.259  
- **R**: ~0.4967  

La **Regresin Lineal** tanto con la **Ecuacin Normal** como con **SVD** tiene un desempeo similar. El R de aproximadamente **0.497** indica que el modelo explica un poco menos del 50% de la varianza en los datos. Este modelo es sencillo, fcil de interpretar y rpido de entrenar, pero puede ser limitado al no capturar relaciones no lineales complejas entre las variables.

Es adecuado si la relacin entre las caractersticas y el objetivo es ms lineal, lo cual puede no ser el caso aqu, dado que estamos trabajando con predicciones relacionadas con precios.

## 2. Regresin Polinomial
- **MSE**: 0.00567  
- **R**: 0.989  

La **Regresin Polinomial** de grado 2 tiene un desempeo sobresaliente con un R cercano a 1, lo que indica que el modelo captura la mayora de la varianza de los datos. Al ser capaz de modelar relaciones no lineales, este modelo es ms adecuado para problemas donde las caractersticas y el objetivo tienen una relacin ms compleja. Sin embargo, el riesgo es el sobreajuste si el grado del polinomio es demasiado alto, lo que puede hacer que el modelo se adapte demasiado a los datos de entrenamiento y no generalice bien a nuevos datos.

## 3. Batch Gradient Descent
- **MSE**: 0.3567  
- **R**: 0.307  

El modelo de **Batch Gradient Descent** muestra un rendimiento inferior en comparacin con la regresin lineal y polinomial. El MSE es ms alto y el R es ms bajo, lo que indica que el modelo no est ajustando adecuadamente los datos. Sin embargo, puede ser til cuando se tienen grandes cantidades de datos y se quiere optimizar el tiempo de entrenamiento.

## 4. Stochastic Gradient Descent (SGD)
- **MSE**: 0.2637  
- **R**: 0.488  

El modelo de **Stochastic Gradient Descent (SGD)** tiene un rendimiento mejor que el Batch Gradient Descent pero an es inferior a los modelos de regresin lineal. A pesar de ser ms eficiente en trminos computacionales, especialmente cuando se trabaja con conjuntos de datos grandes, la calidad de las predicciones no es tan buena como con otros mtodos.

## 5. Lasso Regression
- **MSE**: 0.5149  
- **R**: -3.468e-05  

La **Lasso Regression**, que incluye regularizacin L1, muestra un desempeo deficiente en este caso. El R negativo sugiere que el modelo no se ajusta bien a los datos y est sobreajustado. Este tipo de regularizacin puede ayudar en la seleccin de caractersticas, pero en este caso parece haber reducido demasiado la complejidad del modelo, lo que afecta su capacidad para aprender de los datos.

## 6. Ridge Regression
- **MSE**: 0.2592  
- **R**: 0.4967  

La **Ridge Regression** es similar a la regresin lineal pero con regularizacin L2, que ayuda a prevenir el sobreajuste. El rendimiento es casi idntico al de la regresin lineal estndar, lo que sugiere que no es necesario aplicar regularizacin en este caso especfico. Sin embargo, en escenarios con muchas variables y riesgo de sobreajuste, la regularizacin puede ser beneficiosa.

## Conclusin
En general, **la regresin polinomial** ofrece el mejor desempeo en este caso, con un MSE muy bajo y un R cercano a 1. Esto indica que puede capturar de manera efectiva la relacin no lineal entre las caractersticas y el objetivo. Los modelos de **regresin lineal** y **ridge regression** tambin ofrecen un rendimiento decente, mientras que los mtodos de **gradient descent** y **lasso regression** tienen un desempeo inferior en este conjunto de datos.
