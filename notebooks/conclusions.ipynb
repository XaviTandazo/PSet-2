```python
import matplotlib.pyplot as plt
```

```python
modelos = ['Linear Regression (Normal Equation)', 'Linear Regression (SVD)', 'Polynomial Regression', 
           'Batch Gradient Descent', 'SGDRegressor', 'Lasso Regression', 'Ridge Regression']
mse_values = [0.2591627613278307, 0.25916276132783084, 0.0056728202281362446, 0.3566843080910934,
              0.2637288588446325, 0.5149366954573091, 0.2591608210105241]
r2_values = [0.49669201916022154, 0.49669201916022143, 0.9889830788958194, 0.30729994547529427,
             0.4878244129126097, -3.467752484720421e-05, 0.49669578736048614]


# Crear gráfico para MSE
plt.figure(figsize=(12, 6))
plt.barh(modelos, mse_values, color='lightcoral')
plt.xlabel('MSE (Mean Squared Error)')
plt.title('Comparación de MSE entre los Modelos')
plt.show()
# Crear gráfico para R²
plt.figure(figsize=(12, 6))
plt.barh(modelos, r2_values, color='lightblue')
plt.xlabel('R² (Coeficiente de Determinación)')
plt.title('Comparación de R² entre los Modelos')
plt.show()
```

# Análisis de Modelos para la Predicción

## 1. Regresión Lineal (Ecuación Normal y SVD)
- **MSE**: 0.259  
- **R²**: ~0.4967  

La **Regresión Lineal** tanto con la **Ecuación Normal** como con **SVD** tiene un desempeño similar. El R² de aproximadamente **0.497** indica que el modelo explica un poco menos del 50% de la varianza en los datos. Este modelo es sencillo, fácil de interpretar y rápido de entrenar, pero puede ser limitado al no capturar relaciones no lineales complejas entre las variables.

Es adecuado si la relación entre las características y el objetivo es más lineal, lo cual puede no ser el caso aquí, dado que estamos trabajando con predicciones relacionadas con precios.

## 2. Regresión Polinomial
- **MSE**: 0.00567  
- **R²**: 0.989  

La **Regresión Polinomial** de grado 2 tiene un desempeño sobresaliente con un R² cercano a 1, lo que indica que el modelo captura la mayoría de la varianza de los datos. Al ser capaz de modelar relaciones no lineales, este modelo es más adecuado para problemas donde las características y el objetivo tienen una relación más compleja. Sin embargo, el riesgo es el sobreajuste si el grado del polinomio es demasiado alto, lo que puede hacer que el modelo se adapte demasiado a los datos de entrenamiento y no generalice bien a nuevos datos.

## 3. Batch Gradient Descent
- **MSE**: 0.3567  
- **R²**: 0.307  

El modelo de **Batch Gradient Descent** muestra un rendimiento inferior en comparación con la regresión lineal y polinomial. El MSE es más alto y el R² es más bajo, lo que indica que el modelo no está ajustando adecuadamente los datos. Sin embargo, puede ser útil cuando se tienen grandes cantidades de datos y se quiere optimizar el tiempo de entrenamiento.

## 4. Stochastic Gradient Descent (SGD)
- **MSE**: 0.2637  
- **R²**: 0.488  

El modelo de **Stochastic Gradient Descent (SGD)** tiene un rendimiento mejor que el Batch Gradient Descent pero aún es inferior a los modelos de regresión lineal. A pesar de ser más eficiente en términos computacionales, especialmente cuando se trabaja con conjuntos de datos grandes, la calidad de las predicciones no es tan buena como con otros métodos.

## 5. Lasso Regression
- **MSE**: 0.5149  
- **R²**: -3.468e-05  

La **Lasso Regression**, que incluye regularización L1, muestra un desempeño deficiente en este caso. El R² negativo sugiere que el modelo no se ajusta bien a los datos y está sobreajustado. Este tipo de regularización puede ayudar en la selección de características, pero en este caso parece haber reducido demasiado la complejidad del modelo, lo que afecta su capacidad para aprender de los datos.

## 6. Ridge Regression
- **MSE**: 0.2592  
- **R²**: 0.4967  

La **Ridge Regression** es similar a la regresión lineal pero con regularización L2, que ayuda a prevenir el sobreajuste. El rendimiento es casi idéntico al de la regresión lineal estándar, lo que sugiere que no es necesario aplicar regularización en este caso específico. Sin embargo, en escenarios con muchas variables y riesgo de sobreajuste, la regularización puede ser beneficiosa.

## Conclusión
En general, **la regresión polinomial** ofrece el mejor desempeño en este caso, con un MSE muy bajo y un R² cercano a 1. Esto indica que puede capturar de manera efectiva la relación no lineal entre las características y el objetivo. Los modelos de **regresión lineal** y **ridge regression** también ofrecen un rendimiento decente, mientras que los métodos de **gradient descent** y **lasso regression** tienen un desempeño inferior en este conjunto de datos.
